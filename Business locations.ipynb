```python
data2 = data.loc[(data['City'] == 'San Francisco') & data['Business End Date'].isnull() & data['Location End Date'].isnull() & ((data['NAICS Code Description'] == 'Accommodations') | (data['NAICS Code Description'] == 'Food Services'))][['Street Address', 'Source Zipcode']].values
```

```python
data2.shape
```

```python
data2 = data2[:100]
```

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json?address=%s,%s&key=AIzaSyAJ4HW37Q9ip-SGiEXfsk6eTRNm3DQGY0c'
locations = [requests.get(url%(street, int(zipcode))).json()['results'][0]['geometry']['location'].values() for street, zipcode in data2]
```
